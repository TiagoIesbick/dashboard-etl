````python
%%capture
!pip install tabula-py[jpype]
import re
import pandas as pd
from tabula import read_pdf
from datetime import datetime
import pathlib
from google.colab import drive


# mount drive
drive.mount('/content/drive')
````

````python
def str_to_float(value: str) -> float:
  value = str(value).replace('.', '').replace(',', '.')
  value = re.sub(r'[A-Za-z]|\s', '', value)
  return float(0) if value == '' else float(value)

def agrupamento_format(value: str) -> str:
  if re.search(r'^0{2}', value[:2]):
    value = '   ' + value
  elif re.search(r'^[1-9]\d', value[:2]):
    value = '      ' + value
  return value
````

````python
folder = pathlib.Path(r'/content/drive/MyDrive/PARAMETRIZACÃßOÃÉES FOPAG/disponibilidades/PREV')

df_all = pd.DataFrame()

for arquivo in folder.iterdir():
  dates = read_pdf(arquivo, area=[100, 71, 100+16, 71+390], pages=1, pandas_options={'header': None})[0]
  fonte = read_pdf(arquivo, area=[115, 71, 115+20, 71+245], pages=1, pandas_options={'header': None})[0]
  fonte = re.search(r'[\d\.]*', fonte.iloc[0,1]).group()
  date_range = datetime.strptime(dates.iloc[0,3], '%d/%m/%Y') - datetime.strptime(dates.iloc[0,1], '%d/%m/%Y')
  until_end_year_range = datetime(datetime.now().year, 12, 31) - datetime.strptime(dates.iloc[0,3], '%d/%m/%Y')
  df = read_pdf(arquivo, area=[180, 71, 180+370, 71+755], guess=False, pandas_options={'header': [0]}, pages='all', multiple_tables=False, encoding='latin1')[0]
  df.drop(df.loc[df['Agrupamento']=='Agrupamento'].index, inplace=True)
  df_out = df.loc[df['%'].isna()]
  for i in df_out.index:
    df.loc[i-1, 'Agrupamento'] += ' ' + df.loc[i, 'Agrupamento']
  df.drop(df_out.index, inplace=True)
  if len(df.loc[df['Agrupamento']=='Total']) > 1:
    df.drop(df.loc[df['Agrupamento']=='Total'].index[1:], inplace=True)
  df.reset_index(drop=True, inplace=True)
  df.iloc[:, 1:] = df.iloc[:, 1:].map(str_to_float)
  df['Liquidado por dia'] = df['Liquidado'] / date_range.days
  df['Proje√ß√£o Liquida√ß√£o at√© o final do ano, conforme dias corridos'] = (df['Liquidado por dia'] * until_end_year_range.days) + df['Liquidado']
  df['Necessidade ou sobra de recursos projetada, conforme dias corridos'] = df['Atualizado'] - df['Proje√ß√£o Liquida√ß√£o at√© o final do ano, conforme dias corridos']
  df.iloc[:,-3:] = df.iloc[:,-3:].map(lambda x: float('{:.2f}'.format(x)))
  df['unidorc'] = int(re.sub(r'\D', '', df.iloc[1,0]))
  df['fonte'] = fonte
  li = list(df.loc[df['Agrupamento'].str.startswith('00')].index)
  li.append(df.index[-1])
  for i,j in enumerate(li):
    subacao = int(re.sub(r'\D', '',df.loc[j, 'Agrupamento']))
    if i+1 < len(li):
      idx = li[i]
      while idx <= li[i+1]:
        df.loc[idx, 'subacao'] = subacao
        idx += 1
  df.loc[~df['Agrupamento'].str.startswith(('0', 'T')), 'elemento'] = df['Agrupamento'].str.replace(r'\D', '', regex=True)
  df_all = pd.concat([df_all, df], ignore_index=True)

df_all['Agrupamento'] = df_all['Agrupamento'].map(agrupamento_format)

with pd.ExcelWriter(r'/content/drive/MyDrive/cubo/cubo.xlsx') as writer:
  for i in df_all['unidorc'].unique():
    df_all.loc[df_all['unidorc'] == i].to_excel(writer, sheet_name=str(i), index=False)
````
