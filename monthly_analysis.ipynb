```python
import pandas as pd
```

```python
def filter_window(df, date_col='DATE', months=12, extra_years=(1,2)):
    today = pd.Timestamp.today()
    cutoff = today.to_period('M').to_timestamp() - pd.DateOffset(months=1)
    start = cutoff - pd.DateOffset(months=months-1)

    mask_main = (df[date_col] >= start) & (df[date_col] <= cutoff)
    extra_dates = [cutoff - pd.DateOffset(years=y) for y in extra_years]

    return df[mask_main | df[date_col].isin(extra_dates)]

def group_cols(df, cols_to_delete=None):
    cols_to_delete = set(cols_to_delete or [])
    return [c for c in df.columns if c not in cols_to_delete]
```

```python
df_exp = pd.read_csv(r'/content/drive/MyDrive/Dashboard_data/final_data/df_exp.csv', sep=';', parse_dates=['Comp.pagto.'])
df_rev = pd.read_csv(r'/content/drive/MyDrive/Dashboard_data/final_data/df_rev.csv', sep=";", parse_dates=['Data'])
df_exp = filter_window(df_exp, date_col='Comp.pagto.')
df_rev = filter_window(df_rev, date_col='Data')
```

```python
df_rev.groupby(by=group_cols(df_rev, ['valor_arrecadado']), as_index=False).sum()
```

```python
df_exp.groupby(by=group_cols(df_exp, ['Result. pago']), as_index=False).sum()
```
